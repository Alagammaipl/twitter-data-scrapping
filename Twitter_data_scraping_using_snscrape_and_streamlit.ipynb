```python
%%writefile app.py
#importing needed modules

import streamlit as st
import numpy as np
import pandas as pd
import datetime
import pandas as pd
from pymongo import MongoClient
import json

#initiallizing mongodb connection

client=MongoClient("mongodb+srv://Alagammai:alagammai@cluster0.tgilzlf.mongodb.net/?retryWrites=true&w=majority")
db=client['twitter_scrape']
# fuction to write data in database

@st.cache
def load_mongo_data(coll_name):
    data = db.coll_name.find()
    for documents in data:
        print(documents)
# function to convert data to csv and json formats

def convert_df(df):
    return df.to_csv().encode('utf-8')
    return df.to_json().encode('utf-8')
#creating dataframe and performing cleaning inside the app to be accessed in app

df = pd.read_json("/content/user-tweets.json",  lines=True)
df['date'] = pd.to_datetime(df['date']).dt.date
df['time'] = pd.to_datetime(df['date']).dt.time
df1 = df.drop(['_type','time', 'conversationId','sourceUrl','links','media','rawContent','quoteCount', 'inReplyToTweetId','inReplyToUser','mentionedUsers','coordinates', 'retweetedTweet', 'quotedTweet','sourceLabel','place',	'hashtags',	'cashtags',	'card','viewCount', 'vibe'], axis = 1)

#getting keywords, tweet cout limit, date range from user

keyword = st.text_input('keyword', 'key')
tweetlimit = st.slider('maximum no of times tweeted', 0, 258512, 100)
start_date = st.date_input("select date range",datetime.date(2023, 1, 1))
end_date = st.date_input("select date range",datetime.date(2023, 1, 24))

# filterting data based on user input and displaying them 

df_filtered = df1[df1['renderedContent'].str.contains(keyword)]
df2= df_filtered[df_filtered['retweetCount'] <= tweetlimit]
df3 = df2[df2['date'] >= start_date]
df4 = df3[df3['date'] <= end_date]
st.write(df4)

# converting datetime format to str as it will be needed for adding collection to mongodb database

df4['date']=df4['date'].astype(str)

#uploading data in database

if st.button('upload'):
    my_col=db["your data"]
    my_col.insert_many(df4.to_dict('records'))
    convert = convert_df(df4)

    # creating download options for user

    st.download_button(
      label="Download data as CSV",
      data=convert,
      file_name='yourdata.csv',
      mime='text/csv',)
    st.download_button(label="Download data as json",data=convert,file_name='yourdata.json',mime='text/json')
```
